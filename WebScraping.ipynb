```python
import requests
from bs4 import BeautifulSoup as bs
import re
import pandas as pd
```

```python
url = 'https://www.carandbike.com/research/top-10-best-cars-in-the-world'
```

```python
r = requests.get(url)
```

```python
print(r)
```

```python
soup = bs(r.content)
```

```python
print(soup)
```

```python
s = soup.find('table')
s
```

```python
content = s.find_all('td')
content
```

```python
content[0]
pattern = '\n|<td>|<p>|</p>|</td>|&gt;|&lt;'
re.sub(pattern,'',str(content[7]))
```

```python
Name_of_the_Car = []
Top_Speed = []
zero_100_km_per_h = []
Torque = []
Power =  []
```

```python
count = 0
for i in content:
  pattern = '\n|<td>|<p>|</p>|</td>|&gt;|&lt;'
  value = re.sub(pattern,'',str(i))

  if count == 0:
    Name_of_the_Car.append(value)
    count +=1
  elif count == 1:
    Top_Speed.append(value)
    count +=1
  elif count == 2:
    zero_100_km_per_h.append(value)
    count +=1
  elif count == 3:
    Torque.append(value)
    count +=1
  else :
    Power.append(value)
    count = 0
```

```python
Name_of_the_Car = Name_of_the_Car[1:]
Top_Speed = Top_Speed[1:]
zero_100_km_per_h = zero_100_km_per_h[1:]
Torque = Torque[1:]
Power =  Power[1:]
```

```python
len(Power)
```

```python
# Create a dictionary containing the data
data = {'Name': Name_of_the_Car[0:10],
        'Top_Speed': Top_Speed[0:10],
        '0-100_km_per_h': zero_100_km_per_h[0:10],
        'Torque':Torque[0:10],
        'Power':Power[0:10]}

# Create a DataFrame from the dictionary
df = pd.DataFrame(data)
```
